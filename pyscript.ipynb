### Q1. What is the average median income of the data set and check the distribution of data using appropriate plots. Please explain the distribution of the plot.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Question 1: Average median income and its distribution
avg_median_income = housing_data['median_income'].mean()
# Size of the VIZ Figure
plt.figure(figsize=(10, 6))
#Using Histogram with KDE
sns.histplot(housing_data['median_income'], kde=True)
plt.title('Distribution of Median Income')
plt.xlabel('Median Income')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

print(f'The average Median Income is {avg_median_income}')
```

Used histogram with KDE to shows the distribution of median income across the dataset. The distribution appears right-skewed, shows that a majority of the areas have a lower median income, with fewer areas having a higher median income.

### 2. Draw an appropriate plot to see the distribution of housing_median_age and explain your observations

```python
# Size of the VIZ Figure
plt.figure(figsize=(10, 6))
#Using Histogram with KDE ,Bins is for Histogram no of Distributions is needed on y axis.
sns.histplot(housing_data['housing_median_age'], kde=True, bins=30)
plt.title('Distribution of Housing Median Age')
plt.xlabel('Housing Median Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Used histogram with KDE to show distribution of housing_median_age. The distribution shows a high frequency of areas with a median age around 15 years, and then again a noticeable peak around 50 years. This indicates that there are significant numbers of housing areas that are relatively new (around 15 years old) and quite old (around 50 years old), with fewer areas in the intermediate age range.

### 3. Show with the help of visualization, how median_income and median_house_values are related?

```python
# Size of the VIZ Figure
plt.figure(figsize=(10, 6))
#Using ScatterPlot, s is used to determine size of the plot point
sns.scatterplot(x='median_income', y='median_house_value', data=housing_data,s=3)
plt.title('Relationship between Median Income and Median House Value')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.grid(True)
plt.show()
```

As the median income increases, the median house value tends to increase as well. This suggests that higher-income areas are associated with more expensive housing. However, it is notable that there are horizontal lines at certain house values, indicating price capped at value 500000.

### 5. Create a data set by filling the missing data with the mean value of the total_bedrooms in the original data set.

```python

# Question 5: Create a dataset by filling missing total_bedrooms with the mean value

# Calculate the mean value of total_bedrooms
mean_total_bedrooms = housing_data['total_bedrooms'].mean()

# Creating a new dataset by replacing NaN values in total_bedrooms with the mean value
dataset_with_mean_bedrooms = housing_data.copy()
dataset_with_mean_bedrooms['total_bedrooms'].fillna(mean_total_bedrooms, inplace=True)

# Display the first few rows of the new dataset
dataset_with_mean_bedrooms.head()
```

Created a new dataset where missing total_bedrooms data is replaced with the mean value of total_bedrooms from the original dataset. The first few rows of this updated dataset are displayed above using head function.

### 6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.

```python
def calculate_median(dataframe, column_name):
    """
    Function to calculate the median value of a specified column in a DataFrame.

    """
    return dataframe[column_name].median()

## Calling Function can pass any Column of the Dataset
median_income_median = calculate_median(housing_data, 'median_income')
print(f'The median of the field Provided above is {median_income_median}')

```

Created a function named calculate_median. This function calculates the median value of a specified column in a DataFrame.
As an example, the median of median_income using this function is approximately 3.53

### Q 10 Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
def categorize_bedrooms(bedrooms):
    if bedrooms <= 10:
        return 'Small'
    elif 11 <= bedrooms < 1000:
        return 'Medium'
    else:
        return 'Large'

# Apply the categorization function to the dataset
housing_data['total_bedroom_size'] = housing_data['total_bedrooms'].apply(categorize_bedrooms)


# Display the first few rows of the dataset with the new column
housing_data.head(100)

```
